```python
def get_key_information(file):
    with open(file, 'r') as f:
        key = f.readline().strip()
        name = f.readline().strip()
        return ( key, name )
```

```python
range_inclusive = lambda start, end : range(start, end + 1)
ranges = {v:k for k, v in {
    'management' : (2, 31),
    'operations' : (32, 47),
    'finance' : (48, 57),
    'computers' : (58, 75),
    'engineering' : (76, 96),
    'science' : (97, 118),
    'social_services' : (119, 126),
    'law' : (127, 131),
    'education' : (132, 142),
    'media_entertainment' : (143, 160),
    'health_practitioner' : (161, 192),
    'health_support' : (193, 203),
    'protective' : (204, 221),
    'food_preparation' : (222, 234),
    'building_maintenance' : (235, 240),
    'service' : (241, 260),
    'sales_retail' : (261, 278),
    'admin_support' : (279, 330),
    'farm_fish_forest' : (331, 338),
    'construction' : (339, 368),
    'extraction' : (369, 376),
    'maintenance' : (377, 412),
    'production' : (413, 490),
    'transportation' : (491, 526),
}.items()}
ranges
```

```python
data = {}
for xpose in transposed:
    with open(xpose,'r') as f:
        name = xpose[:17]
        csv = csvkit.reader(f)
        for row in [r for r in csv if r[0].find('_') > -1]:
            [kind, key] = row[0].split('_')
            if key not in data:
                data[key] = {}
            data[key]['name'] = row[1]
            if name not in data[key]:
                data[key][name] = {}
            if(kind == 'HD01'):
                data[key][name]['total'] = row[2]
            if(kind == 'HD02'):
                data[key][name]['margin'] = row[2]
print data
```

```python
data_1yr = {}
data_3yr = {}
data_5yr = {}
for xpose in transposed:
    with open(xpose, 'r') as f:
        name = xpose[11:17]
        year = int(xpose[7:8])
        if year == 1:
            d = data_1yr
        elif year == 3:
            d = data_3yr
        elif year == 5:
            d = data_5yr
        csv = csvkit.reader(f)
        for row in [r for r in csv if r[0].find('_') > -1]:
            [kind, key] = row[0].split('_')
            if key not in d:
                row_name = row[1].split(' - ')[1] if row[1].find('-') > -1 else row[1]
                d[key] = { 'name' : row_name }
                key_id = int(key[2:])
                for k,v in ranges.items():
                    if key_id in range_inclusive(*k):
                        d[key]['group'] = v
            if name not in d[key]:
                d[key][name] = {}
            if kind == 'HD01':
                d[key][name]['total'] = row[2]
            elif kind == 'HD02':
                d[key][name]['margin'] = row[2]

data_1yr
```

```python
def calculate_group(data, group_name):
    B24121 = 0
    B24122 = 0
    B24123 = 0
    
    B24124 = 0
    B24125 = 0
    B24126 = 0
    
    group = [prof for prof in data.values() if 'group' in prof and prof['group'] == group_name]
    # print group

    for i in group:
        B24124 += int(i['B24124']['total'])
        B24125 += int(i['B24125']['total'])
        B24126 += int(i['B24126']['total'])
        
        B24121 += int(i['B24121']['total']) * int(i['B24124']['total'])
        B24122 += int(i['B24122']['total']) * int(i['B24125']['total'])
        try:
            B24123 += int(i['B24123']['total']) * int(i['B24126']['total'])
        except ValueError:
            print i['B24123']['total']
    
    return {
        'name' : group_name,
        'group' : group_name,
        'B24124' : { 'total' : B24124 },
        'B24125' : { 'total' : B24125 },
        'B24126' : { 'total' : B24126 },

        'B24121' : { 'total' : B24121 / B24124 },
        'B24122' : { 'total' : B24122 / B24125 },
        'B24123' : { 'total' : B24123 / B24126 },
    }
```

```python
data_array_5yr = {
    'professions' : [dict(v.items() + {'key' : k}.items()) for k,v in data_5yr.items()]
}
data_array_5yr['groups'] = [calculate_group(data_5yr, v) for v in ranges.values()]
data_array_5yr
```

```python
with open('data_5yr.json', 'w') as f:
    f.write(json.dumps(data_array_5yr, sort_keys=True, indent=2))
```
